```python
def prime(n):
    if n>1 :
        for i in range(2,n):
            if n%i==0:
                print('Not a Prime Number')
                break
        else :
             print('Prime Number')
    else :
             print('Not a Prime Number')
```

```python
prime(4)
```

```python
prime(3)
```

```python
prime(5)
```

```python
prime(6)
```

```python
prime(7)
```

## LAMBDA FUNCTIONS / ANONYMOUS FUNCTIONS 

```python
def sq(n):
    return n*n
```

```python
sq(2)
```

```python
a1 = lambda n : n*n
```

```python
a1(5)
```

```python
sum = lambda a,b : a+b 
```

```python
sum(6,5)
```

using if else (cant use loops)

```python
def even(n):
    if n%2==0:
        print('even')
    else :
        print('odd')
```

```python
even(6)
```

```python
a = lambda n : 'even' if n%2==0 else 'odd'
```

```python
a(2)
```

```python
def even_list(b1):
    even = []
    for i in b1:
        if i%2 == 0:
            even.append(i)
    return even
```

```python
a1 = [2,3,4,5,6,7,8,9]
even_list(a1)
```
