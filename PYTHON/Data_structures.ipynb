# if-else condition

```python
age = 22

if age>=18:
    print('You are allowed to drive')
```

```python
age = 16

if age>=18:
    print('You are allowed to drive')
```

```python
age = 22

if age>=18:
    print('You are allowed to drive')
    print('Done')
```

```python
age = 22

if age>=18:
    print('You are allowed to drive')
print('Done')
```

```python
age = 15

if age>=18:
    print('You are allowed to drive')
    print('Done')
```

```python
age = 15

if age>=18:
    print('You are allowed to drive')
print('Done')
```

```python
age = 16

if age>=18:
    print('You are allowed to drive')
else:
    print('You are not allowed to drive')
```

```python
age = 25

if age>=18:
    print('You are allowed to drive')
else:
    print('You are not allowed to drive')
```

```python
age = 18

if age>18:
    print('You are allowed to drive')
elif age==18:
    print('Take a test drive')
else:
    print('You are not allowed to drive')
```
