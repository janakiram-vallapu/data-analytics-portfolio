## LOOPS

## While loop

```python
i = 1
while i<=10 :
    print (i)
    i = i+1 
```

```python
i=1
while(i<=10) :
    if ( i != 10 ):
        print( '5 x', i ,' =', i*5)
    else :
        print( '5 x', i ,'=', i*5)
    
    i=i+1
```

## BREAK

```python
i = 1
while i<=10 :
    print(i)
    if i == 5 :
        break
    i = i+1
```

## CONTINUE

```python
i = 0
while i<10 :
    i = i+1
    if i==5:
        continue
    print(i)
    
```

Write a python code to create otp verification system

```python
otp = 2639
max_attempts = 3
count = 1 
while ( count <= max_attempts):
    user_otp=int(input('enter the otp'))
    if(user_otp == otp):
        print('correct otp')
        break
    else :
        print ('incorrect otp')
        print( 'attempts left = ',max_attempts - count)
        count = count+1
else :
    print('OTP expired')
```
