## If else

```python
if age>18 :
    print ("eligible for vote ")
else :
    print ("not eligible for vote ")
```

## elif

```python
age = int(input('enter the age'))
if age>18 :
    print('you are eligible for test drive')
elif age==18 :
    print('take a test drive')
else :
    print('not eligible')
    
```

## nested if else

```python
num = int(input('Enter the number'))
if num%2==0 :
    print('even number')
    if num%4==0 :
        print('divisible by 4')
    else  :
        print ('not divisible by 4')
else :
     print('odd number')
```
